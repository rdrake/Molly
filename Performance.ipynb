```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

from pandas import DataFrame
from pandas.io.parsers import read_csv
```

```python
data = read_csv("2013-09-20-14-05-21-result.csv", index_col="idx")
data = data.sort(["hops", "method"])
```

```python
methods = set([method for method in data.method])

plt.figure()

for method in methods:
    ax = data[data.method==method].plot(x="hops", y="duration", label=method)

ax.set_xlabel("Hops")
ax.set_ylabel("Time (ms)")

plt.legend(loc="best")
```

```python
df1 = read_csv("2013-09-20-15-57-15-result.csv")

from pandas.tools.plotting import bootstrap_plot

bootstrap_plot(df1.duration, size=3, samples=30, color="grey")
```
